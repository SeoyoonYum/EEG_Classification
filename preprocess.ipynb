```python
raw_list = [
        'Pinch_Both_01_EPOCX_184303_2023.07.14T15.12.31+09.00.edf',
        'Pinch_Both_02_EPOCX_184303_2023.07.14T15.14.26+09.00.edf',
        'Pinch_Both_03_EPOCX_184303_2023.07.14T15.16.05+09.00.edf',
        'Pinch_Both_04_EPOCX_184303_2023.07.14T15.19.35+09.00.edf',
        'Pinch_Both_05_EPOCX_184303_2023.07.14T15.21.34+09.00.edf'
       ]

event_list = [
         'Pinch_Both_01_EPOCX_184303_2023.07.14T15.12.31+09.00_intervalMarker.csv',
        'Pinch_Both_02_EPOCX_184303_2023.07.14T15.14.26+09.00_intervalMarker.csv',
        'Pinch_Both_03_EPOCX_184303_2023.07.14T15.16.05+09.00_intervalMarker.csv',
        'Pinch_Both_04_EPOCX_184303_2023.07.14T15.19.35+09.00_intervalMarker.csv',
        'Pinch_Both_05_EPOCX_184303_2023.07.14T15.21.34+09.00_intervalMarker.csv'
        ]
```

```python
filtered_markers = ['1','2','3','4','5']
epoch_tmin = -2
epoch_tmax = 2
picked_channels = ['AF4','F8','F4','FC6','T8','P8','O2']
for i in range(len(raw_list)):
    raw_path = raw_list[i]
    event_path = event_list[i]
    raw = mne.io.read_raw_edf(raw_path,preload=True)
    event = pd.read_csv(event_path)
    event = event[event['marker_value'].isin(filtered_markers)]
    event = event[['marker_value','latency']]
    
    interval = np.diff(raw.times)[0]
    stim_data = np.zeros((1,len(raw.times)))
    j=0
    for k in range(len(raw.times)):
        if(j==len(event)):
            break
        if(abs(raw.times[k]-event.iloc[j,1])<interval/2):
            stim_data[0,k]=event.iloc[j,0]
            j+=1
    
    info = mne.create_info(['STI'],raw.info['sfreq'], ['stim'])
    stim_raw = mne.io.RawArray(stim_data, info)
    raw.add_channels([stim_raw], force_update_info=True)
    
    raw.pick_channels(['AF3',
     'F7',
     'F3',
     'FC5',
     'T7',
     'P7',
     'O1',
     'O2',
     'P8',
     'T8',
     'FC6',
     'F4',
     'F8',
     'AF4',
     'STI'])
    
    raw.set_montage('easycap-M1')
    raw.plot_sensors(ch_type='eeg')
    ica = mne.preprocessing.ICA(method='fastica')
    
    
    raw_filtered = raw.copy().filter(8,None)
    events = mne.find_events(raw_filtered,stim_channel='STI')
    # event_dict = {
    #     'finger_movement':1
    # }
    ica.fit(raw_filtered)
    
    ic_labels = mne_icalabel.label_components(raw_filtered,ica,'iclabel')
    labels = ic_labels["labels"]
    print(f"ICA components and labels: {list(zip(range(len(labels)), labels))}")
    exclude_idx = [idx for idx, label in enumerate(labels) if label not in ["brain"]]
    print(f"Excluding these ICA components: {exclude_idx}")
    
    # ica.apply() changes the Raw object in-place, so let's make a copy first:
    reconst_raw = raw_filtered.copy()
    ica.apply(reconst_raw, exclude=exclude_idx)
    #ica.apply(reconst_raw,include=[7])\
    
    event_dict = {
    'finger_movement':1
    }
    
    epochs = mne.Epochs(
    reconst_raw,
    events,
    event_dict,
    tmin=epoch_tmin,
    tmax=epoch_tmax,
    #reject=dict(eeg=150e-6),
    #flat=dict(eeg=1e-6),
    baseline=None,
    preload=True,
    picks=picked_channels,
    #picks=['eeg']
    ) 
    epochs.save(f'../epochs/pinch/epoch{i}-epo.fif')
```
